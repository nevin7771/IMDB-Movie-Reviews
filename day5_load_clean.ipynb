# Week 1 — Day 5: IMDB load & clean

**Scope:** load reviews from folders → `(texts, labels)` → clean with regex → validate.

**Not today:** TF-IDF, model training, Streamlit.

Run this notebook from the project root (same folder as `aclImdb/`) so paths resolve.

```python
import os
import re
from tqdm.auto import tqdm


def load_data(path):
    texts = []
    labels = []

    for label in ["pos", "neg"]:
        folder = os.path.join(path, label)
        files = [f for f in os.listdir(folder) if f.endswith(".txt")]

        for file in tqdm(files, desc=f"{os.path.basename(path)}/{label}"):
            fp = os.path.join(folder, file)
            with open(fp, encoding="utf-8") as f:
                texts.append(f.read())
                labels.append(1 if label == "pos" else 0)

    return texts, labels


def clean_text(text):
    text = text.lower()
    text = re.sub(r"<.*?>", "", text)  # remove HTML tags
    text = re.sub(r"[^a-zA-Z]", " ", text)  # keep only letters
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
DATA_ROOT = "aclImdb"

train_texts, train_labels = load_data(os.path.join(DATA_ROOT, "train"))
test_texts, test_labels = load_data(os.path.join(DATA_ROOT, "test"))

print("Train size:", len(train_texts))
print("Test size:", len(test_texts))
```

```python
train_texts = [clean_text(t) for t in tqdm(train_texts, desc="clean train")]
test_texts = [clean_text(t) for t in tqdm(test_texts, desc="clean test")]

print("\nSample cleaned review (first 300 chars):\n", train_texts[0][:300])
```

```python
# Validation — don’t skip
print("First 10 train labels:", train_labels[:10])
print("Label set:", set(train_labels))  # expect {0, 1}
assert set(train_labels) == {0, 1}
assert len(train_texts) == len(train_labels) == 25_000
assert len(test_texts) == len(test_labels) == 25_000
print("Checks passed.")
```
